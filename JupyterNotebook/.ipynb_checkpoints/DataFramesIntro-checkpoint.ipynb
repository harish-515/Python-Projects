Fetching the values from the dataframe

1. Label Based Indexing
    df10.loc[range of rows,range of columns]
    ex : df10.loc["735 Dolores St":"332 Hill St","Country":"ID"]
2. Postion Based Indexing
    df10.iloc[range of rows,range of columns]
    ex : df10.iloc[1:3,1:3]
